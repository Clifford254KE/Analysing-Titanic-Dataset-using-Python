```python
Titanic_df.drop(columns=['deck'], inplace=True)
```

```python
print(Titanic_df.isnull().sum())
```

```python
Titanic_df['age'].fillna(Titanic_df['age'].median(), inplace=True)
```

```python
print(Titanic_df.isnull().sum())
```

```python
sns.countplot(x="survived", data=Titanic_df, legend=True,palette="viridis")
plt.ylabel("count")
plt.title("People who Survived or did not Survive")
plt.show()
```

```python
sns.countplot(x="pclass", data=Titanic_df, legend=True,palette="viridis")
plt.ylabel("count")
plt.title("Distribution of people per classes")
plt.show()
```

```python
first_class_count=(Titanic_df['pclass']==1).sum()
print("First class: ", first_class_count)
second_class_count=(Titanic_df['pclass']==2).sum()
print("Second class: ", second_class_count)
third_class_count=(Titanic_df['pclass']==3).sum()
print("Third class: ", third_class_count)
```

```python
labels=["First class","Second class","Third class"]
sizes=[first_class_count,second_class_count,third_class_count]
plt.pie(sizes,labels=labels)
plt.title("Class Distribution")
plt.show()
```

```python
labels=["First class","Second class","Third class"]
sizes=[first_class_count,second_class_count,third_class_count]
plt.pie(sizes,labels=labels, autopct='%1.2f%%')
plt.title("Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="pclass", hue="survived", data=Titanic_df)
plt.title('Survived per class')
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(Titanic_df.pclass [Titanic_df.survived==0], label= "died")
sns.kdeplot(Titanic_df.pclass[Titanic_df.survived==1], label= "survived")
plt.title("Survived per Class")
plt.xlabel("classes")
plt.ylabel("Normalized count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(data=Titanic_df[Titanic_df["survived"] == 0], x="pclass", label="Died", shade=True)

sns.kdeplot(data=Titanic_df[Titanic_df["survived"] == 1], x="pclass", label="Survived", shade=True)

plt.title("Survived per Class")
plt.xlabel("classes")
plt.ylabel("Normalized count")
plt.legend()
plt.show()
```
